**Решение для функции на промежутке [-10, 11]**

```python
from sympy import *
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = symbols('x')
```

```python
f = -12 * x ** 4 * sin(cos(x)) - 18 * x ** 3 + 5 * x ** 2 + 10 * x - 30
f
```

**График функции на промежутке [-10, 11]**

```python
plot(f, (x, -10 , 11))
```

```python
plot(f, (x, -2.5, 3))
```

*Вариант 1: поиск примерных значений, с помощью цикла*

(Если 2 соседних числа имеют разные знаки, то они близки к корню уравнения)

```python
roots = []
for i in np.arange(-10, 11, 0.001):
    f1 = -12 * i ** 4 * sin(cos(i)) - 18 * i ** 3 + 5 * i ** 2 + 10 * i - 30
    f2 = -12 * (i-0.001) ** 4 * sin(cos(i-0.001)) - 18 * (i-0.001) ** 3 + 5 * (i-0.001) ** 2 + 10 * (i-0.001) - 30
    if f1 * f2 < 0:
        roots.append(round(i, 2))
roots
```

*Вариант 2: По графику найдены примерные промежутки, в которых находятся корни, затем ищутся корни с помощью nsolve*

```python
intervals = [(-8, -7.5), (-5.2, -5), (-1.5, -1), (2, 2.5), (4, 5), (8, 9), (10, 11)]
roots = [nsolve(f, x, i) for i in intervals]
roots
```
